## What neighborhoods seem to be the most popular?

```python
popular = pd.read_csv('popular.csv')

popular
```

Assuming the number of days a unit is available out of 365 is an inverse indicator of the unit's popularity (and not an indicator of some other factor), the most popular districts appear to be Westlake, Montlake, Yesler Terrace, South Delridge, Arbor Heights, Windermere, Roxhill, Wedgewood, Atlantic, Lower Queen Anne, and North Beach/Blue Ridge. There is a large grouping of AirBnB units with no listed neighborhood, which are low in average availability. The lowest grouping is the Industrial District, but it has only 2 AirBnB properties, so it is unlikely the most popular.
